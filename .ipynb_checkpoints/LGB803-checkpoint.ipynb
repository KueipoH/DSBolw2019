```python
import numpy as np
import pandas as pd
import gc
import time
from contextlib import contextmanager
from boostaroota import BoostARoota
import shap

from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.model_selection import KFold, StratifiedKFold
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import cmath
from tqdm import tqdm
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python

@contextmanager
def timer(title):
    t0 = time.time()
    yield
    print("{} - done in {:.0f}s".format(title, time.time() - t0))
    
def drop_nan(train_df, NaN_rate = 0.8):
    mis_val = train_df.isnull().sum()
    mis_val_percent = 100 * train_df.isnull().sum() / len(train_df)
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    mis = mis_val_table.rename(
            columns = {0 : 'Missing Values', 1 : 'percentage'})

    miss_list = list(mis[mis['percentage'] > NaN_rate*100].index)
        

    return miss_list

def geometric_mean(x):
    return np.exp(np.log(x[x>0]).mean())


def reduce_mem_usage(df):
    """ iterate through all the columns of a dataframe and modify the data type
        to reduce memory usage.        
    """
    start_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage of dataframe is {:.2f} MB'.format(start_mem))
    
    for i,col in tqdm(enumerate(df.columns),total=len(df.columns)):
        col_type = df[col].dtype
        
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        else:
            df[col] = df[col].astype('object')

    end_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
    print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))
    
    return df


def bins_data(app_train):
    buckets = [-25229,-23813,-19866,-17368,-14018,-13427,-11679,-8609,-7489]
    labels = ['DAYS_BIRTH0', 'DAYS_BIRTH1', 'DAYS_BIRTH2', 'DAYS_BIRTH3', 'DAYS_BIRTH4', 'DAYS_BIRTH5', 'DAYS_BIRTH6', 'DAYS_BIRTH7']
    app_train['DAYS_BIRTH_BINS'] = pd.cut(app_train.DAYS_BIRTH, bins=buckets, labels = labels, include_lowest=True)
    app_train['DAYS_BIRTH_BINS'] = app_train['DAYS_BIRTH_BINS'].astype('object')
    
    buckets = [-17912,-9508,-3924,-3076,-2448,-1405,0]
    labels = ['DAYS_EMPLOYED0', 'DAYS_EMPLOYED1', 'DAYS_EMPLOYED2', 'DAYS_EMPLOYED3', 'DAYS_EMPLOYED4', 'DAYS_EMPLOYE5']
    app_train['DAYS_EMPLOYED_BINS'] = pd.cut(app_train.DAYS_EMPLOYED, bins=buckets, labels = labels, include_lowest=True)
    app_train['DAYS_EMPLOYED_BINS'] = app_train['DAYS_EMPLOYED_BINS'].astype('object')
    
    buckets = [-7197,-4219,-3118,-1887,0]
    labels = ['DAYS_ID_PUBLISH0', 'DAYS_ID_PUBLISH1', 'DAYS_ID_PUBLISH2', 'DAYS_ID_PUBLISH3']
    app_train['DAYS_ID_PUBLISH_BINS'] = pd.cut(app_train.DAYS_ID_PUBLISH, bins=buckets, labels = labels, include_lowest=True)
    app_train['DAYS_ID_PUBLISH_BINS'] = app_train['DAYS_ID_PUBLISH_BINS'].astype('object')
    print ("bin_data done")
    return app_train

def duplicate_columns(frame):
    groups = frame.columns.to_series().groupby(frame.dtypes).groups
    dups = []
    for t, v in groups.items():
        cs = frame[v].columns
        vs = frame[v]
        lcs = len(cs)
        for i in range(lcs):
            ia = vs.iloc[:,i].values
            for j in range(i+1, lcs):
                ja = vs.iloc[:,j].values
                if np.array_equiv(ia, ja):
                    dups.append(cs[i])
                    break
    return dups 

def woe_data(df):
    app_train = df[df['TARGET'].notnull()]
    app_test = df[df['TARGET'].isnull()]
    cols = [col for col in app_train.columns if col not in ["TARGET"] and app_train[col].dtype == 'int64' and 
    len(app_train[col].value_counts()) < 30]
    for col in app_train:
        if app_train[col].dtype == 'object' or col in cols:
            prob_df = app_train.groupby([col])['TARGET'].mean()
            prob_df = pd.DataFrame(prob_df)
            prob_df['UNTARGET'] = 1-prob_df.TARGET
            prob_df.loc[prob_df.TARGET == 0, 'TARGET'] = 0.00001
            prob_df['WoE'] = np.log(prob_df.TARGET/prob_df.UNTARGET)
            ordered_labels = prob_df['WoE'].to_dict()
            app_train[col] = app_train[col].map(ordered_labels)
            app_test[col] = app_test[col].map(ordered_labels)
            app_train[col].replace([np.inf,-np.inf],app_train[col].median(),inplace=True)
            app_test[col].replace([np.inf,-np.inf],app_train[col].median(),inplace=True)

            
    return app_train.append(app_test).reset_index()

def fill_missing(df):
    cat_cols = [col for col in df.columns if df[col].dtype == 'object']
    num_cols = [col for col in df.columns if col not in cat_cols]
    for f_ in cat_cols:
        most_frequency_category = df.groupby([f_])[f_].count().sort_values(ascending=False).index[0]
        df[f_].fillna(most_frequency_category,inplace=True)
    for f_ in num_cols:
        df[f_].fillna(df[f_].median(),inplace=True)
    return df 

def one_hot_encoder(df, nan_as_category = True):
    original_columns = list(df.columns)
    categorical_columns = [col for col in df.columns if df[col].dtype == 'object']
    df = pd.get_dummies(df, columns= categorical_columns, dummy_na= nan_as_category)
    new_columns = [c for c in df.columns if c not in original_columns]
    return df, new_columns

# Preprocess application_train.csv and application_test.csv
def application_train_test(num_rows = None, nan_as_category = True):
    # Read data and merge
    df = pd.read_csv('../input/application_train.csv', nrows= num_rows)
#     df = reduce_mem_usage(df)
    test_df = pd.read_csv('../input/application_test.csv', nrows= num_rows)
#     test_df = reduce_mem_usage(test_df)
    print("Train samples: {}, test samples: {}".format(len(df), len(test_df)))
    df = df.append(test_df).reset_index()
    df['DAYS_EMPLOYED'].replace(365243, np.nan, inplace= True)

    cat_features = [f_ for f_ in df.columns if df[f_].dtype == 'object']
    for f_ in cat_features:
        df.loc[df[f_] == "XNA",f_] = np.nan
        df.loc[df[f_] == "XAP",f_] = np.nan
    df['BEFORE_MISSING'] = df.isnull().mean(axis=1)
    for f_ in cat_features:
        if df[f_].isnull().values.any():
            most_frequency_category = df.groupby([f_])[f_].count().sort_values(ascending=False).index[0]
            df[f_+"_na"] = np.where(df[f_].isnull(), 1, 0)
            df[f_].fillna(most_frequency_category,inplace=True)  
    num_features = [f_ for f_ in df.columns if f_ not in cat_features and f_ not in ['TARGET']]
    for f_ in num_features:
        if df[f_].isnull().values.any():
            df[f_+"_na"] = np.where(df[f_].isnull(), 1, 0)
            df[f_].fillna(df[f_].median(),inplace=True)
    df = bins_data(df)
    df = woe_data(df)
    # Some simple new features (percentages)
    # External sources
    df['EXT_SOURCE_WEIGHTS'] = df.EXT_SOURCE_1 * 2 + df.EXT_SOURCE_2 * 3 + df.EXT_SOURCE_3 * 4
    for function_name in ['min', 'max', 'sum', 'mean', 'nanmedian']:
        df['EXT_SOURCE_{}'.format(function_name.upper())] = eval('np.{}'.format(function_name))(df[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']], axis=1)
    df['EXT_SOURCE_123'] = df['EXT_SOURCE_1']*df['EXT_SOURCE_2']*df['EXT_SOURCE_3']
    df['EXT_SOURCE_1 EXT_SOURCE_3'] = df['EXT_SOURCE_1'] * df['EXT_SOURCE_3']
    df['EXT_SOURCE_3^2'] = df['EXT_SOURCE_3'] * df['EXT_SOURCE_3'] 
    df['EXT_SOURCE_1^3'] = df['EXT_SOURCE_1'] * df['EXT_SOURCE_1'] * df['EXT_SOURCE_1']
    df['EXT_SOURCE_3^2 DAYS_BIRTH'] = df['EXT_SOURCE_3'] * df['EXT_SOURCE_3'] * df['DAYS_BIRTH']
    df['EXT_SOURCE_1_PERC'] = df['EXT_SOURCE_1'] / df['EXT_SOURCE_SUM']
    df['EXT_SOURCE_2_PERC'] = df['EXT_SOURCE_2'] / df['EXT_SOURCE_SUM']
    df['EXT_SOURCE_3_PERC'] = df['EXT_SOURCE_3'] / df['EXT_SOURCE_SUM']
    df['EXT_SOURCE_STD'] = df[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].std(axis=1)
    
    
    df['NEW_EXT_SOURCES_GEO'] = df[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].apply(geometric_mean, axis=1)
    df['NEW_EXT_SOURCES_PROD'] = df[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].prod(axis=1, skipna=True, min_count=1)

    ##################################################################        
    for col in cat_features:
        df[col+'_SALARY'] = df.groupby(col).AMT_INCOME_TOTAL.transform('mean')
        df[col+'_SALARY_RATIO'] = df['AMT_INCOME_TOTAL']/df[col+'_SALARY']
        df[col+'_PAID'] = df.groupby(col).AMT_ANNUITY.transform('mean')
        df[col+'_PAID_RATIO'] = df['AMT_ANNUITY']/df[col+'_PAID']
        df[col+'_LEND'] = df.groupby(col).AMT_CREDIT.transform('mean')
        df[col+'_LEND_RATIO'] = df['AMT_CREDIT']/df[col+'_LEND']
        df[col+'_GOOD'] = df.groupby(col).AMT_GOODS_PRICE.transform('mean')
        df[col+'_GOOD_RATIO'] = df['AMT_GOODS_PRICE']/df[col+'_GOOD']
        del df[col+'_SALARY'],df[col+'_PAID'],df[col+'_LEND'],df[col+'_GOOD'];gc.collect()


        df[col+'_DAYS_BIRTH'] = df.groupby(col).DAYS_BIRTH.transform('mean')
        df[col+'_DAYS_BIRTH_RATIO'] = df['DAYS_BIRTH']/df[col+'_DAYS_BIRTH']
        df[col+'_DAYS_EMPLOYED'] = df.groupby(col).DAYS_EMPLOYED.transform('mean')
        df[col+'_DAYS_EMPLOYED_RATIO'] = df['DAYS_EMPLOYED']/df[col+'_DAYS_EMPLOYED']
        df[col+'_DAYS_LAST_PHONE_CHANGE'] = df.groupby(col).DAYS_LAST_PHONE_CHANGE.transform('mean')
        df[col+'_DAYS_LAST_PHONE_CHANGE_RATIO'] = df['DAYS_LAST_PHONE_CHANGE']/df[col+'_DAYS_LAST_PHONE_CHANGE']
        df[col+'_DAYS_ID_PUBLISH'] = df.groupby(col).DAYS_ID_PUBLISH.transform('mean')
        df[col+'_DAYS_ID_PUBLISH_RATIO'] = df['DAYS_ID_PUBLISH']/df[col+'_DAYS_ID_PUBLISH']
        df[col+'_DAYS_REGISTRATION'] = df.groupby(col).DAYS_REGISTRATION.transform('mean')
        df[col+'_DAYS_REGISTRATION_RATIO'] = df['DAYS_REGISTRATION']/df[col+'_DAYS_REGISTRATION']
        del df[col+'_DAYS_BIRTH'],df[col+'_DAYS_EMPLOYED'],df[col+'_DAYS_LAST_PHONE_CHANGE'],df[col+'_DAYS_ID_PUBLISH'],df[col+'_DAYS_REGISTRATION'];gc.collect()


        df[col+'_1'] = df.groupby(col).EXT_SOURCE_1.transform('mean')
        df[col+'_1_RATIO'] = df['EXT_SOURCE_1']/df[col+'_1']
        df[col+'_2'] = df.groupby(col).EXT_SOURCE_2.transform('mean')
        df[col+'_2_RATIO'] = df['EXT_SOURCE_2']/df[col+'_2']
        df[col+'_3'] = df.groupby(col).EXT_SOURCE_3.transform('mean')
        df[col+'_3_RATIO'] = df['EXT_SOURCE_3']/df[col+'_3']
        df[col+'_MEAN'] = df.groupby(col).EXT_SOURCE_MEAN.transform('mean')
        df[col+'_MEAN_RATIO'] = df['EXT_SOURCE_MEAN']/df[col+'_MEAN']
        df[col+'_SUM'] = df.groupby(col).EXT_SOURCE_SUM.transform('mean')
        df[col+'_SUM_RATIO'] = df['EXT_SOURCE_SUM']/df[col+'_SUM']
        del df[col+'_1'],df[col+'_2'],df[col+'_3'],df[col+'_MEAN'],df[col+'_SUM'];gc.collect()

    ##################################################################
    df['MEAN_BUILDING_SCORE_AVG'] = df.iloc[:, 44:58].mean(skipna=True, axis=1)
    df['TOTAL_BUILDING_SCORE_AVG'] = df.iloc[:, 44:58].sum(skipna=True, axis=1)
    df['FLAG_DOCUMENT_TOTAL'] = df.iloc[:, 96:116].sum(axis=1)
    df['AMT_REQ_CREDIT_BUREAU_TOTAL'] = df.iloc[:, 116:122].sum(axis=1)
    df['long_employment'] = (df['DAYS_EMPLOYED'] > -2000).astype(int)
    df['retirement_age'] = (df['DAYS_BIRTH'] > -14000).astype(int)

    docs = [_f for _f in df.columns if 'FLAG_DOC' in _f]
    live = [_f for _f in df.columns if ('FLAG_' in _f) & ('FLAG_DOC' not in _f) & ('_FLAG_' not in _f)]
    df['NEW_DOC_IND_KURT'] = df[docs].kurtosis(axis=1)
    df['NEW_LIVE_IND_SUM'] = df[live].sum(axis=1)

    inc_by_org = df[['AMT_INCOME_TOTAL', 'ORGANIZATION_TYPE']].groupby('ORGANIZATION_TYPE').median()['AMT_INCOME_TOTAL']
    df['NEW_INC_BY_ORG'] = df['ORGANIZATION_TYPE'].map(inc_by_org)

    df['NEW_CAR_TO_BIRTH_RATIO'] = df['OWN_CAR_AGE'] / df['DAYS_BIRTH']
    df['NEW_CAR_TO_EMPLOY_RATIO'] = df['OWN_CAR_AGE'] / df['DAYS_EMPLOYED']
    df['NEW_PHONE_TO_BIRTH_RATIO'] = df['DAYS_LAST_PHONE_CHANGE'] / df['DAYS_BIRTH']
    df['NEW_PHONE_TO_BIRTH_RATIO'] = df['DAYS_LAST_PHONE_CHANGE'] / df['DAYS_EMPLOYED']
    df['children_ratio'] = df['CNT_CHILDREN'] / df['CNT_FAM_MEMBERS']

    
    ####################################################
#     df['NEW_CREDIT_TO_ANNUITY_RATIO_Square'] = df['AMT_CREDIT']**2 / df['AMT_ANNUITY']

    ####################################################
    
    df['NEW_CREDIT_TO_ANNUITY_RATIO'] = df['AMT_CREDIT'] / df['AMT_ANNUITY']
    df['NEW_CREDIT_TO_GOODS_RATIO'] = df['AMT_CREDIT'] / df['AMT_GOODS_PRICE']
    df['NEW_CREDIT_TO_INCOME_RATIO'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']
    df['SAVING_MONEY'] = (df['AMT_INCOME_TOTAL']*df['DAYS_EMPLOYED'])/(-365)
    df['NEW_CREDIT_TO_SAVING_RATIO'] = df['AMT_CREDIT'] / df['SAVING_MONEY']
    df['NEW_ANNUITY_TO_SAVING_RATIO'] = df['AMT_ANNUITY'] / df['SAVING_MONEY']
    df['SAVING_MONEY2'] = (df['AMT_INCOME_TOTAL']*df['DAYS_BIRTH'])/(-365)
    df['NEW_CREDIT_TO_SAVING_RATIO2'] = df['AMT_CREDIT'] / df['SAVING_MONEY2']
    df['NEW_ANNUITY_TO_SAVING_RATIO2'] = df['AMT_ANNUITY'] / df['SAVING_MONEY2']
    df['SAVING_MONEY3'] = (df['AMT_INCOME_TOTAL']*df['DAYS_ID_PUBLISH'])/(-365)
    df['NEW_CREDIT_TO_SAVING_RATIO3'] = df['AMT_CREDIT'] / df['SAVING_MONEY3']
    df['NEW_ANNUITY_TO_SAVING_RATIO3'] = df['AMT_ANNUITY'] / df['SAVING_MONEY3']
    del df['SAVING_MONEY'],df['SAVING_MONEY2'],df['SAVING_MONEY3'];gc.collect()

    df['NEW_ANNUITY_TO_INCOME_RATIO'] = df['AMT_ANNUITY'] / (1 + df['AMT_INCOME_TOTAL'])
    df['NEW_INC_PER_CHLD'] = df['AMT_INCOME_TOTAL'] / (1 + df['CNT_CHILDREN'])
    df['NEW_EMPLOY_TO_BIRTH_RATIO'] = df['DAYS_EMPLOYED'] / df['DAYS_BIRTH']
    average = 0
    feats = [f for f in df.columns if f not in ['TARGET','SK_ID_CURR','SK_ID_BUREAU','SK_ID_PREV', 'index']]
    for col in feats:
        df[col].fillna(df[col].median(axis=0), inplace=True)
        average += np.array(df[col].rank(ascending=1))
    average = average/len(feats)
    df['APPLICATION_AVERAGE'] = average
    del test_df
    gc.collect()
    return df

# Preprocess bureau.csv and bureau_balance.csv
def bureau_and_balance(num_rows = None, nan_as_category = True):
    bureau = pd.read_csv('../input/bureau.csv', nrows = num_rows)
#     bureau = reduce_mem_usage(bureau)
    bureau['BUREAU_MISSING'] = bureau.isnull().mean(axis=1)
    #
    #bureau['PREV_MISSING_CNT'] = (bureau==np.nan).sum(axis=1).astype(float)
    #
    bureau = fill_missing(bureau)
    bureau['AMT_CREDIT_SUM_DEBT_PERC'] = bureau['AMT_CREDIT_SUM_DEBT'] / bureau['AMT_CREDIT_SUM']
    bureau['OVERDUE_ANNUITY'] = bureau['AMT_CREDIT_SUM_OVERDUE']*(-365) / bureau['CREDIT_DAY_OVERDUE']
    bureau['DAYS_CREDIT_UPDATE_DIFF'] = bureau['DAYS_CREDIT_UPDATE'] - bureau['DAYS_CREDIT']
    bureau['CREDIT_DAY_OVERDUE_DIFF'] = bureau['CREDIT_DAY_OVERDUE'] - bureau['DAYS_CREDIT']
    bureau_new = []

    for col in ['CREDIT_TYPE','CREDIT_CURRENCY','CREDIT_ACTIVE','CNT_CREDIT_PROLONG']:
        bureau[col+'_DAYS_CREDIT'] = bureau.groupby(col).DAYS_CREDIT.transform('mean')
        bureau[col+'_DAYS_CREDIT_RATIO'] = bureau['DAYS_CREDIT']/bureau[col+'_DAYS_CREDIT']
        bureau_new.append(col+'_DAYS_CREDIT_RATIO')
        bureau[col+'_DAYS_CREDIT_ENDDATE'] = bureau.groupby(col).DAYS_CREDIT_ENDDATE.transform('mean')
        bureau[col+'_DAYS_CREDIT_ENDDATE_RATIO'] = bureau['DAYS_CREDIT_ENDDATE']/bureau[col+'_DAYS_CREDIT_ENDDATE']
        bureau_new.append(col+'_DAYS_CREDIT_ENDDATE_RATIO')
        bureau[col+'_AMT_CREDIT_SUM'] = bureau.groupby(col).AMT_CREDIT_SUM.transform('mean')
        bureau[col+'_AMT_CREDIT_SUM_RATIO'] = bureau['AMT_CREDIT_SUM']/bureau[col+'_AMT_CREDIT_SUM']
        bureau_new.append(col+'_AMT_CREDIT_SUM_RATIO')
        bureau[col+'_AMT_CREDIT_SUM_DEBT'] = bureau.groupby(col).AMT_CREDIT_SUM_DEBT.transform('mean')
        bureau[col+'_AMT_CREDIT_SUM_DEBT_RATIO'] = bureau['AMT_CREDIT_SUM_DEBT']/bureau[col+'_AMT_CREDIT_SUM_DEBT']
        bureau_new.append(col+'_AMT_CREDIT_SUM_DEBT_RATIO')
        bureau[col+'_DAYS_CREDIT_UPDATE'] = bureau.groupby(col).DAYS_CREDIT_UPDATE.transform('mean')
        bureau[col+'_DAYS_CREDIT_UPDATE_RATIO'] = bureau['DAYS_CREDIT_UPDATE']/bureau[col+'_DAYS_CREDIT_UPDATE']
        bureau_new.append(col+'_DAYS_CREDIT_UPDATE_RATIO')

        del bureau[col+'_DAYS_CREDIT'],bureau[col+'_DAYS_CREDIT_ENDDATE'],bureau[col+'_AMT_CREDIT_SUM'],bureau[col+'_AMT_CREDIT_SUM_DEBT'],bureau[col+'_DAYS_CREDIT_UPDATE'];gc.collect()


    bureau, bureau_cat = one_hot_encoder(bureau, nan_as_category)

    bb = pd.read_csv('../input/bureau_balance.csv', nrows = num_rows)
#     bb = reduce_mem_usage(bb)
    bb = fill_missing(bb)
    bb, bb_cat = one_hot_encoder(bb, nan_as_category)
    
    # Bureau balance: Perform aggregations and merge with bureau.csv
    bb_aggregations = {'MONTHS_BALANCE': ['min', 'max', 'size']}
    for col in bb_cat:
        bb_aggregations[col] = ['mean']
    bb_agg = bb.groupby('SK_ID_BUREAU').agg(bb_aggregations)
    bb_agg.columns = pd.Index([e[0] + "_" + e[1].upper() for e in bb_agg.columns.tolist()])
    bureau = bureau.join(bb_agg, how='left', on='SK_ID_BUREAU')
    del bb, bb_agg
    gc.collect()
    average = 0
    feats = [f for f in bureau.columns if f not in ['TARGET','SK_ID_CURR','SK_ID_BUREAU','SK_ID_PREV', 'index']]
    for col in feats:
        bureau[col].fillna(bureau[col].median(axis=0), inplace=True)
        average += np.array(bureau[col].rank(ascending=1))
    average = average/len(feats)
    bureau['BUREAU_AVERAGE'] = average
    cat_aggregations = {}
    for cat in bureau_cat: cat_aggregations[cat] = ['mean', 'nunique']
    new_aggregations = {}
    for cat in bureau_new: new_aggregations[cat] = ['mean']
    num_aggregations = {}
    bureau_num = [f for f in bureau.columns if f not in bureau_cat and f not in bureau_new]
    for num in bureau_num: num_aggregations[num] = ['min', 'max', 'mean', 'sum', 'var']

    bureau_agg = bureau.groupby('SK_ID_CURR').agg({**num_aggregations, **cat_aggregations, **new_aggregations})
    bureau_agg.columns = pd.Index(['BURO_' + e[0] + "_" + e[1].upper() for e in bureau_agg.columns.tolist()])
    # Bureau: Active credits - using only numerical aggregations
    active = bureau[bureau['CREDIT_ACTIVE_Active'] == 1]
    active_agg = active.groupby('SK_ID_CURR').agg(num_aggregations)
    active_agg.columns = pd.Index(['ACT_' + e[0] + "_" + e[1].upper() for e in active_agg.columns.tolist()])
    bureau_agg = bureau_agg.join(active_agg, how='left', on='SK_ID_CURR')
    del active, active_agg
    gc.collect()
    # Bureau: Closed credits - using only numerical aggregations
    closed = bureau[bureau['CREDIT_ACTIVE_Closed'] == 1]
    closed_agg = closed.groupby('SK_ID_CURR').agg(num_aggregations)
    closed_agg.columns = pd.Index(['CLS_' + e[0] + "_" + e[1].upper() for e in closed_agg.columns.tolist()])
    bureau_agg = bureau_agg.join(closed_agg, how='left', on='SK_ID_CURR')
    del closed, closed_agg, bureau
    gc.collect()
    return bureau_agg

# Preprocess previous_applications.csv  
def previous_applications(num_rows = None, nan_as_category = True):
    prev = pd.read_csv('../input/previous_application.csv', nrows = num_rows)
#     prev = reduce_mem_usage(prev)
    # Days 365.243 values -> nan
    prev['DAYS_FIRST_DRAWING'].replace(365243, np.nan, inplace= True)
    prev['DAYS_FIRST_DUE'].replace(365243, np.nan, inplace= True)
    prev['DAYS_LAST_DUE_1ST_VERSION'].replace(365243, np.nan, inplace= True)
    prev['DAYS_LAST_DUE'].replace(365243, np.nan, inplace= True)
    prev['DAYS_TERMINATION'].replace(365243, np.nan, inplace= True)
    prev['PREV_MISSING'] = prev.isnull().mean(axis=1)
    #
    #prev['PREV_MISSING_CNT'] = (prev==np.nan).sum(axis=1).astype(float)
    #
    prev = fill_missing(prev)
    new_cols = ['NAME_CASH_LOAN_PURPOSE','CODE_REJECT_REASON','NAME_GOODS_CATEGORY','CHANNEL_TYPE','NAME_SELLER_INDUSTRY','PRODUCT_COMBINATION']
    for col in new_cols:
        prev[col+'_CREDIT'] = prev.groupby(col).AMT_CREDIT.transform('mean')
        prev[col+'_CREDIT_RATIO'] = prev['AMT_CREDIT']/prev[col+'_CREDIT']
        del prev[col+'_CREDIT'];gc.collect()
    new_cols = []

    cat_features = [f_ for f_ in prev.columns if prev[f_].dtype == 'object']
    for col in cat_features:
        prev[col+'_AMT_APPLICATION'] = prev.groupby(col).AMT_APPLICATION.transform('mean')
        prev[col+'_AMT_APPLICATION_RATIO'] = prev['AMT_APPLICATION']/prev[col+'_AMT_APPLICATION']
        new_cols.append(col+'_AMT_APPLICATION_RATIO')
        prev[col+'_AMT_CREDIT'] = prev.groupby(col).AMT_CREDIT.transform('mean')
        prev[col+'_AMT_CREDIT_RATIO'] = prev['AMT_CREDIT']/prev[col+'_AMT_CREDIT']
        new_cols.append(col+'_AMT_CREDIT_RATIO')
        prev[col+'_AMT_ANNUITY'] = prev.groupby(col).AMT_ANNUITY.transform('mean')
        prev[col+'_AMT_ANNUITY_RATIO'] = prev['AMT_ANNUITY']/prev[col+'_AMT_ANNUITY']
        new_cols.append(col+'_AMT_ANNUITY_RATIO')
        prev[col+'_AMT_GOODS_PRICE'] = prev.groupby(col).AMT_GOODS_PRICE.transform('mean')
        prev[col+'_AMT_GOODS_PRICE_RATIO'] = prev['AMT_GOODS_PRICE']/prev[col+'_AMT_GOODS_PRICE']
        new_cols.append(col+'_AMT_GOODS_PRICE_RATIO')
        prev[col+'_DAYS_DECISION'] = prev.groupby(col).DAYS_DECISION.transform('mean')
        prev[col+'_DAYS_DECISION_RATIO'] = prev['DAYS_DECISION']/prev[col+'_DAYS_DECISION']
        new_cols.append(col+'_DAYS_DECISION_RATIO')
        del prev[col+'_AMT_APPLICATION'],prev[col+'_AMT_CREDIT'],prev[col+'_AMT_ANNUITY'],prev[col+'_AMT_GOODS_PRICE'],prev[col+'_DAYS_DECISION'];gc.collect()

    prev, cat_cols = one_hot_encoder(prev, nan_as_category= True)
    # Add feature: value ask / value received percentage
    prev['APP_CREDIT_PERC'] = prev['AMT_CREDIT']/prev['AMT_APPLICATION']
    prev['APP_CREDIT_DIFF'] = prev['AMT_APPLICATION']-prev['AMT_CREDIT']
    prev['AMT_CREDIT_DAYS'] = prev['AMT_CREDIT']*prev['DAYS_DECISION']
    prev['AMT_APPLICATION_DAYS'] = prev['AMT_APPLICATION']*prev['DAYS_DECISION']

    prev['AMT_ANNUITY_YEARS_APPLICATION'] = prev['AMT_APPLICATION']/prev['AMT_ANNUITY']
    prev['AMT_ANNUITY_YEARS_CREDIT'] = prev['AMT_CREDIT']/prev['AMT_ANNUITY']

    prev['AMT_DOWN_PAYMENT_PERC_APPLICATION'] = prev['AMT_DOWN_PAYMENT']/prev['AMT_APPLICATION']
    prev['AMT_DOWN_PAYMENT_PERC_CREDIT'] = prev['AMT_DOWN_PAYMENT']/prev['AMT_CREDIT']

    prev['AMT_GOODS_PRICE_PERC_APPLICATION'] = prev['AMT_GOODS_PRICE']/prev['AMT_APPLICATION']
    prev['AMT_GOODS_PRICE_PERC_CREDIT'] = prev['AMT_GOODS_PRICE']/prev['AMT_CREDIT']

    prev['DAYS_FIRST_DUE_DIFF'] = prev['DAYS_LAST_DUE_1ST_VERSION']-prev['DAYS_FIRST_DUE']
    prev['DAYS_LAST_DUE_DIFF'] = prev['DAYS_LAST_DUE']-prev['DAYS_LAST_DUE_1ST_VERSION']
    prev['DAYS_TERMINATION_DIFF'] = prev['DAYS_TERMINATION']-prev['DAYS_FIRST_DRAWING']

    prev['RATE_PREC_PRIMARY'] = prev['RATE_DOWN_PAYMENT']/prev['RATE_INTEREST_PRIMARY']
    prev['RATE_PREC_PRIVILEGED'] = prev['RATE_DOWN_PAYMENT']/prev['RATE_INTEREST_PRIVILEGED']

    average = 0
    feats = [f for f in prev.columns if f not in ['TARGET','SK_ID_CURR','SK_ID_BUREAU','SK_ID_PREV', 'index']]
    for col in feats:
        prev[col].fillna(prev[col].median(axis=0), inplace=True)
        average += np.array(prev[col].rank(ascending=1))
    average = average/len(feats)
    prev['PREV_AVERAGE'] = average

    cat_aggregations = {}
    for cat in cat_cols: cat_aggregations[cat] = ['mean', 'nunique']
    new_aggregations = {}
    for cat in new_cols: new_aggregations[cat] = ['mean']
    num_aggregations = {}
    num_cols = [f for f in prev.columns if f not in cat_cols and f not in new_cols]
    for num in num_cols: num_aggregations[num] = ['min', 'max', 'mean', 'sum', 'var']


    prev_agg = prev.groupby('SK_ID_CURR').agg({**num_aggregations, **cat_aggregations, **new_aggregations})
    prev_agg.columns = pd.Index(['PREV_' + e[0] + "_" + e[1].upper() for e in prev_agg.columns.tolist()])
    # Previous Applications: Approved Applications - only numerical features
    approved = prev[prev['NAME_CONTRACT_STATUS_Approved'] == 1]
    approved_agg = approved.groupby('SK_ID_CURR').agg(num_aggregations)
    approved_agg.columns = pd.Index(['APR_' + e[0] + "_" + e[1].upper() for e in approved_agg.columns.tolist()])
    prev_agg = prev_agg.join(approved_agg, how='left', on='SK_ID_CURR')
    # Previous Applications: Refused Applications - only numerical features
    refused = prev[prev['NAME_CONTRACT_STATUS_Refused'] == 1]
    refused_agg = refused.groupby('SK_ID_CURR').agg(num_aggregations)
    refused_agg.columns = pd.Index(['REF_' + e[0] + "_" + e[1].upper() for e in refused_agg.columns.tolist()])
    prev_agg = prev_agg.join(refused_agg, how='left', on='SK_ID_CURR')
    del refused, refused_agg, approved, approved_agg, prev
    gc.collect()
    return prev_agg

# Preprocess POS_CASH_balance.csv
def pos_cash(num_rows = None, nan_as_category = True):
    pos = pd.read_csv('../input/POS_CASH_balance.csv', nrows = num_rows)
#     pos = reduce_mem_usage(pos)
    pos = fill_missing(pos)

    for col in ['NAME_CONTRACT_STATUS']:
        pos[col+'_MONTHS_BALANCE'] = pos.groupby(col).MONTHS_BALANCE.transform('mean')
        pos[col+'_MONTHS_BALANCE_RATIO'] = pos['MONTHS_BALANCE']/pos[col+'_MONTHS_BALANCE']
        pos[col+'_SK_DPD'] = pos.groupby(col).SK_DPD.transform('mean')
        pos[col+'_SK_DPD_RATIO'] = pos['SK_DPD']/pos[col+'_SK_DPD']
        pos[col+'_SK_DPD_DEF'] = pos.groupby(col).SK_DPD_DEF.transform('mean')
        pos[col+'_SK_DPD_DEF_RATIO'] = pos['SK_DPD_DEF']/pos[col+'_SK_DPD_DEF']
        del pos[col+'_MONTHS_BALANCE'],pos[col+'_SK_DPD'],pos[col+'_SK_DPD_DEF'];gc.collect()

    pos, cat_cols = one_hot_encoder(pos, nan_as_category= True)
    # categorical_feats = [f for f in pos.columns if pos[f].dtype == 'object']
    # for f_ in categorical_feats:
    #     pos[f_], _ = pd.factorize(pos[f_])
    # Features
    pos['CNT_INSTALMENT_SUM'] = pos['CNT_INSTALMENT'] + pos['CNT_INSTALMENT_FUTURE']
    pos['CNT_INSTALMENT_MINUS'] = pos['CNT_INSTALMENT_FUTURE'] - pos['CNT_INSTALMENT']
    pos['SK_DPD_SUM'] = pos['SK_DPD'] + pos['SK_DPD_DEF']
    pos['SK_DPD_MINUS'] = pos['SK_DPD'] - pos['SK_DPD_DEF']
    average = 0
    feats = [f for f in pos.columns if f not in ['TARGET','SK_ID_CURR','SK_ID_BUREAU','SK_ID_PREV', 'index']]
    for col in feats:
        pos[col].fillna(pos[col].median(axis=0), inplace=True)
        average += np.array(pos[col].rank(ascending=1))
    average = average/len(feats)
    pos['POS_AVERAGE'] = average
   
    aggregations = {}
    for cat in cat_cols:
        aggregations[cat] = ['mean', 'nunique']
    num_cols = [f for f in pos.columns if f not in cat_cols]
    for num in num_cols:
        aggregations[num] = ['min', 'max', 'mean', 'sum', 'var']

    pos_agg = pos.groupby('SK_ID_CURR').agg(aggregations)
    pos_agg.columns = pd.Index(['POS_' + e[0] + "_" + e[1].upper() for e in pos_agg.columns.tolist()])
    # Count pos cash accounts
    pos_agg['POS_COUNT'] = pos.groupby('SK_ID_CURR').size()
    del pos
    gc.collect()
    return pos_agg
    
# Preprocess installments_payments.csv
def installments_payments(num_rows = None, nan_as_category = True):
    ins = pd.read_csv('../input/installments_payments.csv', nrows = num_rows)
#     ins = reduce_mem_usage(ins)
    ins = fill_missing(ins)
    ins, cat_cols = one_hot_encoder(ins, nan_as_category= True)
    # categorical_feats = [f for f in ins.columns if ins[f].dtype == 'object']
    # for f_ in categorical_feats:
    #     ins[f_], _ = pd.factorize(ins[f_])
    # Percentage and difference paid in each installment (amount paid and installment value)
    ins['PAYMENT_PERC'] = ins['AMT_PAYMENT'] / ins['AMT_INSTALMENT']
    ins['PAYMENT_DIFF'] = ins['AMT_INSTALMENT'] - ins['AMT_PAYMENT']
    
    ins['PAYMENT_MEAN'] = (ins['AMT_INSTALMENT'] + ins['AMT_PAYMENT'])/2

    # Days past due and days before due (no negative values)
    ins['DPD'] = ins['DAYS_ENTRY_PAYMENT'] - ins['DAYS_INSTALMENT']
    ins['DBD'] = ins['DAYS_INSTALMENT'] - ins['DAYS_ENTRY_PAYMENT']
    ins['DPD'] = ins['DPD'].apply(lambda x: x if x > 0 else 0)
    ins['DBD'] = ins['DBD'].apply(lambda x: x if x > 0 else 0)
    ins['ANNUITY_AMT_PAYMENT'] = ins['AMT_PAYMENT']*(-365) / ins['DAYS_ENTRY_PAYMENT']
    ins['ANNUITY_AMT_INSTALMENT'] = ins['AMT_INSTALMENT']*(-365) / ins['DAYS_INSTALMENT']
    ins['ANNUITY_AMT_PERC'] = ins['ANNUITY_AMT_PAYMENT'] / ins['ANNUITY_AMT_INSTALMENT']
    ins['ANNUITY_AMT_DIFF'] = ins['ANNUITY_AMT_INSTALMENT'] - ins['ANNUITY_AMT_PAYMENT']
    ins['instalment_paid_late_in_days'] = ins['DAYS_ENTRY_PAYMENT'] - ins['DAYS_INSTALMENT'] 
    ins['instalment_paid_late'] = (ins['instalment_paid_late_in_days'] > 0).astype(int)
    ins['instalment_paid_over_amount'] = ins['AMT_PAYMENT'] - ins['AMT_INSTALMENT']
    ins['instalment_paid_over'] = (ins['instalment_paid_over_amount'] > 0).astype(int)
    # Features: Perform aggregations

    average = 0
    feats = [f for f in ins.columns if f not in ['TARGET','SK_ID_CURR','SK_ID_BUREAU','SK_ID_PREV', 'index']]
    for col in feats:
        ins[col].fillna(ins[col].median(axis=0), inplace=True)
        average += np.array(ins[col].rank(ascending=1))
    average = average/len(feats)
    ins['INS_AVERAGE'] = average

    aggregations = {}
    for cat in cat_cols:
        aggregations[cat] = ['mean', 'nunique']
    num_cols = [f for f in ins.columns if f not in cat_cols]
    for num in num_cols:
        aggregations[num] = ['min', 'max', 'mean', 'sum', 'var']

    ins_agg = ins.groupby('SK_ID_CURR').agg(aggregations)
    ins_agg.columns = pd.Index(['INS_' + e[0] + "_" + e[1].upper() for e in ins_agg.columns.tolist()])
    # Count installments accounts
    ins_agg['INS_COUNT'] = ins.groupby('SK_ID_CURR').size()
    del ins
    gc.collect()
    return ins_agg

    # Preprocess credit_card_balance.csv
def credit_card_balance(num_rows = None, nan_as_category = True):
    cc = pd.read_csv('../input/credit_card_balance.csv', nrows = num_rows)
#     cc = reduce_mem_usage(cc)
    cc = fill_missing(cc)

    for col in ['NAME_CONTRACT_STATUS']:
        cc[col+'_AMT_DRAWINGS_CURRENT'] = cc.groupby(col).AMT_DRAWINGS_CURRENT.transform('mean')
        cc[col+'_AMT_DRAWINGS_CURRENT_RATIO'] = cc['AMT_DRAWINGS_CURRENT']/cc[col+'_AMT_DRAWINGS_CURRENT']
        cc[col+'_SK_DPD'] = cc.groupby(col).SK_DPD.transform('mean')
        cc[col+'_SK_DPD_RATIO'] = cc['SK_DPD']/cc[col+'_SK_DPD']
        cc[col+'_SK_DPD_DEF'] = cc.groupby(col).SK_DPD_DEF.transform('mean')
        cc[col+'_SK_DPD_DEF_RATIO'] = cc['SK_DPD_DEF']/cc[col+'_SK_DPD_DEF']
        del cc[col+'_AMT_DRAWINGS_CURRENT'],cc[col+'_SK_DPD'],cc[col+'_SK_DPD_DEF'];gc.collect()

    cc, cat_cols = one_hot_encoder(cc, nan_as_category= True)
    # categorical_feats = [f for f in cc.columns if cc[f].dtype == 'object']
    # for f_ in categorical_feats:
    #     cc[f_], _ = pd.factorize(cc[f_])
    cc['AMT_DRAWINGS_CURRENT_RATIO']= cc['AMT_DRAWINGS_ATM_CURRENT'] / (cc['AMT_DRAWINGS_CURRENT'] + cc['AMT_DRAWINGS_OTHER_CURRENT'] + cc['AMT_DRAWINGS_POS_CURRENT'])
    cc['SK_DPD_SUM'] = cc['SK_DPD'] + cc['SK_DPD_DEF']
    cc['SK_DPD_MINUS'] = cc['SK_DPD'] - cc['SK_DPD_DEF']
    cc['AMT_DRAWINGS_ATM_CURRENT_PERC'] = cc['AMT_DRAWINGS_ATM_CURRENT'] / cc['AMT_DRAWINGS_CURRENT']
    cc['AMT_DRAWINGS_OTHER_CURRENT_PERC'] = cc['AMT_DRAWINGS_OTHER_CURRENT'] / cc['AMT_DRAWINGS_CURRENT']
    cc['AMT_DRAWINGS_POS_CURRENT_PERC'] = cc['AMT_DRAWINGS_POS_CURRENT'] / cc['AMT_DRAWINGS_CURRENT']
    cc['AMT_DRAWINGS_CURRENT_SUM'] = cc['AMT_DRAWINGS_ATM_CURRENT']+cc['AMT_DRAWINGS_OTHER_CURRENT']+cc['AMT_DRAWINGS_POS_CURRENT']

    cc['AMT_PAYMENT_CURRENT_PERC'] = cc['AMT_PAYMENT_CURRENT'] / cc['AMT_PAYMENT_TOTAL_CURRENT']
    cc['AMT_RECEIVABLE_PRINCIPAL_PERC'] = cc['AMT_RECEIVABLE_PRINCIPAL'] / cc['AMT_TOTAL_RECEIVABLE']
    cc['AMT_TOTAL_RECEIVABLE_SUM'] = cc['AMT_PAYMENT_CURRENT'] + cc['AMT_RECEIVABLE_PRINCIPAL']

    cc['AMT_RECIVABLE_PERC'] = cc['AMT_RECIVABLE'] / cc['AMT_TOTAL_RECEIVABLE']
    cc['CNT_DRAWINGS_SUM'] = cc['CNT_DRAWINGS_ATM_CURRENT']+cc['AMT_DRAWINGS_OTHER_CURRENT']+cc['CNT_DRAWINGS_POS_CURRENT']

    average = 0
    feats = [f for f in cc.columns if f not in ['TARGET','SK_ID_CURR','SK_ID_BUREAU','SK_ID_PREV', 'index']]
    for col in feats:
        cc[col].fillna(cc[col].median(axis=0), inplace=True)
        average += np.array(cc[col].rank(ascending=1))
    average = average/len(feats)
    cc['CC_AVERAGE'] = average

    aggregations = {}
    for cat in cat_cols:
        aggregations[cat] = ['mean', 'nunique']
    num_cols = [f for f in cc.columns if f not in cat_cols]
    for num in num_cols:
        aggregations[num] = ['min', 'max', 'mean', 'sum', 'var']

    cc_agg = cc.groupby('SK_ID_CURR').agg(aggregations)
    cc_agg.columns = pd.Index(['CC_' + e[0] + "_" + e[1].upper() for e in cc_agg.columns.tolist()])
    # Count credit card lines
    cc_agg['CC_COUNT'] = cc.groupby('SK_ID_CURR').size()
    del cc
    gc.collect()
    return cc_agg

def kfold_lightgbm(df, num_folds, stratified = False):

    f_to_clean = [f_ for f_ in df.columns if len(df[f_].value_counts()) < 2]
    print ("Number of excluded feature %d" %len(f_to_clean))
    df.drop(f_to_clean, axis=1, inplace=True)
    print ("number of feature after delete excluded feature : {}".format(len(df.columns)))

    dups = duplicate_columns(df)
    df.drop(dups, axis=1, inplace=True)
    print ("number of feature after delete duplicated feature : {}".format(len(df.columns)))
    average = 0
    
    ######################################
#     dropfeat = drop_nan(df, NaN_rate = 0.9)
#     df.drop(dropfeat, axis=1, inplace=True)
    feats = [f for f in df.columns if f not in ['TARGET','SK_ID_CURR','SK_ID_BUREAU','SK_ID_PREV', 'index']]

    for col in feats:
#         df[col].fillna(df[col].median(axis=0), inplace=True)
#         df[col].replace([np.inf,-np.inf], np.nan, inplace=True)
        average += np.array(df[col].rank(ascending=1))
    average = average/len(feats)
    df['AVERAGE'] = average
    print (df[feats].isnull().any().any())
#     df = reduce_mem_usage(df)
    # Divide in training/validation and test data
    train_df = df[df['TARGET'].notnull()]
    test_df = df[df['TARGET'].isnull()]
    del df
    gc.collect()

#        # read Auto Encoder Features
#     print('Loaging Auto Encoder Features!')
#     train_auto = pd.read_csv('../data/Auto_trn_08_07.csv')
#     test_auto = pd.read_csv('../data/Auto_sub_08_07.csv')

#     print("Before Starting LightGBM. TrainAE shape: {}, testAE shape: {}".format(train_auto.shape, test_auto.shape))

    # concate original data
    print('Original Data concating Auto Encoder Features!')
#     train_df = pd.concat([train_df, train_auto], axis=1)
#     test_df = pd.concat([test_df, test_auto], axis=1)


    feats = [f for f in train_df.columns if f not in ['TARGET','SK_ID_CURR','SK_ID_BUREAU','SK_ID_PREV', 'index']]

    from boostaroota import BoostARoota
    br = BoostARoota(metric='logloss')
    br.fit(train_df[feats], train_df['TARGET'])
    feats = br.keep_vars_
    feats.to_csv('features.csv', index= False)
    feats = feats.tolist()



#     feats = pd.read_csv('features.csv',names=["FEATURES"], header=None)
#     feats = feats['FEATURES'].values.tolist()
    corr_matrix = train_df[feats].corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
    to_drop = [column for column in upper.columns if any(upper[column] > 0.999)]
    print('Before Dropping high corr Train shape: {}, test shape: {}'.format(train_df.shape, test_df.shape))

    train_df.drop(to_drop,axis=1,inplace=True)
    test_df.drop(to_drop,axis=1,inplace=True)
    print('After Dropping high corr Train shape: {}, test shape: {}'.format(train_df.shape, test_df.shape))

    feats.append('TARGET')
    feats.append('SK_ID_CURR')
    drop_feats = [f for f in train_df.columns if f not in feats]
    train_df.drop(drop_feats,axis=1,inplace=True)
    test_df.drop(drop_feats,axis=1,inplace=True)
    print("Train shape: {}, test shape: {}".format(train_df.shape, test_df.shape))

    
    print('Original Data concating Auto Encoder Features!')
    test_df = test_df[test_df['SK_ID_CURR'].notnull()]
    train_df.to_csv('data/train_newfea.csv', index=False)
    test_df.to_csv('data/test_newfea.csv', index=False)  

    print("Starting LightGBM. Train shape: {}, test shape: {}".format(train_df.shape, test_df.shape))

    # Cross validation model
    
    feature_importance_df = pd.DataFrame()
    output = []
    valid = []
    models = []
    
    for i in range(1):
        
        auc_score = []
        if stratified:
            folds = StratifiedKFold(n_splits= num_folds, shuffle=True, random_state=i)
        else:
            folds = KFold(n_splits= num_folds, shuffle=True, random_state=i)
        # Create arrays and dataframes to store results
        oof_preds = np.zeros(train_df.shape[0])
        sub_preds = np.zeros(test_df.shape[0])
        feats = [f for f in train_df.columns if f not in ['TARGET','SK_ID_CURR']]

        for n_fold, (train_idx, valid_idx) in enumerate(folds.split(train_df[feats], train_df['TARGET'])):
            train_x, train_y = train_df[feats].iloc[train_idx], train_df['TARGET'].iloc[train_idx]
            valid_x, valid_y = train_df[feats].iloc[valid_idx], train_df['TARGET'].iloc[valid_idx]

            # LightGBM parameters found by Bayesian optimization
            clf = LGBMClassifier(
                nthread=-1,
                n_estimators=10000,
                learning_rate =  0.013,
                num_leaves=90,
                #boosting_type = "dart",
                feature_fraction = 0.6,
                # subsample_freq = 5,
                bagging_fraction= 0.7,
                max_depth=-1,
                reg_alpha=0.041545473,
                reg_lambda=0.0735294,
                min_split_gain=0.0222415,
                min_child_weight=39.3259775,
                silent=-1,
                verbose=-1,
                random_state = i)

            clf.fit(train_x, train_y, eval_set=[(train_x, train_y), (valid_x, valid_y)], 
                eval_metric= 'auc', verbose= 400, early_stopping_rounds= 100)

            oof_preds[valid_idx] = clf.predict_proba(valid_x, num_iteration=clf.best_iteration_)[:, 1]
            sub_preds += clf.predict_proba(test_df[feats], num_iteration=clf.best_iteration_)[:, 1] / folds.n_splits

            fold_importance_df = pd.DataFrame()
            fold_importance_df["feature"] = feats
            fold_importance_df["importance"] = clf.feature_importances_
            fold_importance_df["shap_values"] = abs(shap.TreeExplainer(clf).shap_values(valid_x)[:,:test_df.shape[1]]).mean(axis=0).T
            fold_importance_df["fold"] = n_fold + 1
            feature_importance_df = pd.concat([feature_importance_df, fold_importance_df], axis=0)
            
            print('Fold %2d AUC : %.6f' % (n_fold + 1, roc_auc_score(valid_y, oof_preds[valid_idx])))
            auc_score.append(roc_auc_score(valid_y, oof_preds[valid_idx]))
            models.append(clf)
            del clf, train_x, train_y, valid_x, valid_y
            gc.collect()
            print('AUC Mean score %6f' % np.mean(auc_score))
            print('AUC STD score %6f' % np.std(auc_score))
        print('Full AUC score %.6f' % roc_auc_score(train_df['TARGET'], oof_preds))

        valid.append(roc_auc_score(train_df['TARGET'], oof_preds))
        output.append(sub_preds)
        # Write submission file and plot feature importance
    print('TOTAL score %.6f' % np.mean(valid, axis=0))
    test_df['TARGET'] = np.mean(output, axis=0)
    test_df[['SK_ID_CURR', 'TARGET']].to_csv(submission_file_name, index= False)
    display_importances(feature_importance_df)
    display_shapley_values(feature_importance_df)

    return feature_importance_df, models, auc_score

    # Display/plot feature importance
def display_importances(feature_importance_df_):
    cols = feature_importance_df_[["feature", "importance"]].groupby("feature").mean().sort_values(by="importance", ascending=False)[:40].index
    best_features = feature_importance_df_.loc[feature_importance_df_.feature.isin(cols)]
    plt.figure(figsize=(8, 10))
    sns.barplot(x="importance", y="feature", data=best_features.sort_values(by="importance", ascending=False))
    plt.title('LightGBM Features (avg over folds)')
    plt.tight_layout()
    plt.savefig('lgbm_importances-01.png')



# Display/plot shapley values
def display_shapley_values(feat_importance):
    best_features = feat_importance[["feature", "shap_values"]].groupby("feature")["shap_values"].agg(['mean', 'std']) \
                                                               .sort_values(by="mean", ascending=False).head(40).reset_index()
    best_features.columns = ["feature", "mean shapley values", "err"]
    plt.figure(figsize=(8, 10))
    sns.barplot(x="mean shapley values", y="feature", xerr=best_features['err'], data=best_features)
    plt.title('LightGBM shapley values (avg over folds)')
    plt.tight_layout()
    plt.show()
```

## Using SHAP (SHapley Additive exPlanations)

```python
# explain the model's predictions using SHAP values
# (same syntax works for LightGBM, CatBoost, and scikit-learn models)
explainer = shap.TreeExplainer(models[0], num_iteration=models[0].best_iteration_)
shap_values = explainer.shap_values(train_df)
```

```python
# Reconstructing the lgb output from shap_values:
def inv_logit(p): return np.exp(p) / (1 + np.exp(p))
base_value = shap_values[0,-1]  ## this old version of shap uses the last index to store the base value
output = base_value + np.sum(shap_values[0,:-1])  ## the remaining indexes are the the linear/additive contributions of each feature

print('Log-odds output:', output, ' Logistic output:', inv_logit(output))
```
